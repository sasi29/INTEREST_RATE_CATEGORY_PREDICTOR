```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('C:/Users/HP-PC/Desktop/file1.csv')
```

```python
dataset['Gender']=pd.get_dummies(dataset['Gender'],drop_first=True)
```

```python
#data=pd.get_dummies(dataset['Income_Verified'],drop_first=True)
dataset['VERIFIED - income source']=data['VERIFIED - income source']
dataset['not verified']=data['not verified']
```

```python
dataset['Home_Owner'].unique()
```

```python
dataset['Home_Owner']=dataset.Home_Owner.str.replace('None','Other')
```

```python
sns.countplot(dataset['Home_Owner'])
```

```python
data=pd.get_dummies(dataset['Home_Owner'])
dataset['Mortgage']=data['Mortgage']
dataset['Rent']=data['Rent']
```

```python
dataset['Purpose_Of_Loan'].unique()
```

```python
sns.countplot(dataset['Purpose_Of_Loan'])
plt.xticks(rotation=90)
```

```python
data=pd.get_dummies(dataset['Purpose_Of_Loan'])
```

```python
data.columns
```

```python
dataset['debt_consolidation']=data['debt_consolidation']
dataset['credit_card']=data['credit_card']
```

```python
dataset['Loan_Amount_Requested']=dataset.Loan_Amount_Requested.str.replace(',','')
```

```python
dataset.columns
```

```python
dataset=dataset.drop(['Home_Owner','Income_Verified','Purpose_Of_Loan'],axis=1)
```
